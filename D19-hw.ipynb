# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()
```

```python
#畫出完整的 cos 圖形
x1 = np.arange(0, 3 * np.pi, 0.1)
y1_cos = np.cos(x1)
plt.plot(x1, y1_cos) # cos 圖形
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
# 顏色由（X，Y）的角度給出。
# 注意標記的大小，顏色和透明度。
n = 100
x = np.random.normal(loc=0, scale=1, size=n)
y = np.random.normal(loc=0, scale=1, size=n)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
cmap = plt.get_cmap("Greys") # 取得灰階的"colormap"
plt.scatter(x, y,c=x+y , cmap=cmap, marker="o", alpha=0.5, edgecolors="r", s=300)
# c=x+y[相加越大, 灰階越深]]
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot")
 # 如果要存成圖形檔:
	# 把 pyplot.show() 換成下面這行:
plt.savefig("filename.png",dpi=300,format="png")
```

    畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？


```python
import pandas as pd
```

```python
data = pd.read_csv("./boston.csv")
```

```python
data
```

```python
data.median()
```

```python
# 畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？
plt.boxplot([data["TAX"],data["B"]]) # TAX & B 欄位的中位數在 300~400 之間
plt.show()
```

```python
# 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
cmap = plt.get_cmap("Purples")
plt.scatter(x=data["NOX"], y=data["DIS"]
, c=data["NOX"].median()*data["NOX"]+data["DIS"].median()*data["DIS"], cmap=cmap, alpha=0.5)
plt.xlabel("NOX")
plt.ylabel("DIS")
plt.title("scatter plot")
plt.show() 
# 可以看出呈現負相關～
```
